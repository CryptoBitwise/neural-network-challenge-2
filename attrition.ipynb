## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
X_df = attrition_df.drop(columns=['Attrition', 'Department'])
X_df.head()


# Create X_df using your selected columns


# Show the data types for X_df
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split


```

```python
# Convert your X data to numeric data types however you see fit
X_df = pd.get_dummies(X_df)
X_df.head()
# Add new code cells as necessary
```

```python
# Create a StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_df)
X_df = scaler.transform(X_df)
X_df



# Fit the StandardScaler to the training data



# Scale the training and testing data
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
enc = OneHotEncoder(sparse_output=False)


# Fit the encoder to the training data
enc.fit(y_df[['Department']])


# Create two new variables by applying the encoder
# to the training and testing data
encoded_department = enc.transform(y_df[['Department']])
encoded_department_df = pd.DataFrame(encoded_department, columns=enc.get_feature_names_out(['Department']))
y_df = y_df.drop(columns=['Department'])
y_df = pd.concat([y_df, encoded_department_df], axis=1)
y_df.head()
```

```python
# Create a OneHotEncoder for the Attrition column
enc = OneHotEncoder(sparse_output=False)


# Fit the encoder to the training data
enc.fit(y_df[['Attrition']])


# Create two new variables by applying the encoder
# to the training and testing data
encoded_attrition = enc.transform(y_df[['Attrition']])
encoded_attrition_df = pd.DataFrame(encoded_attrition, columns=enc.get_feature_names_out(['Attrition']))
y_df = y_df.drop(columns=['Attrition'])
y_df = pd.concat([y_df, encoded_attrition_df], axis=1)
y_df.head()
```

## Part 2: Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data.
X_df.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(X_df.shape[1],))

# Create at least two shared layers
shared_layer_1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer_2 = layers.Dense(32, activation='relu')(shared_layer_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer


# Create the hidden layer
department_hidden = layers.Dense(16, activation='relu')(shared_layer_2)

# Create the output layer
department_output = layers.Dense(3, activation='softmax', name='department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(16, activation='relu')(shared_layer_2)

# Create the output layer
attrition_output = layers.Dense(2, activation='softmax', name='attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])


# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_df, {'attrition_output': y_df[['Attrition_No', 'Attrition_Yes']],
                  'department_output': y_df[['Department_Human Resources', 'Department_Research & Development', 'Department_Sales']]},
          epochs=10, batch_size=32)
```

```python
# Evaluate the model with the testing data
loss, accuracy = model.evaluate(X_df,
                                 {'attrition_output': y_df[['Attrition_No', 'Attrition_Yes']],
                                  'department_output': y_df[['Department_Human Resources', 'Department_Research & Development', 'Department_Sales']]})
print(f"Loss: {loss}, Accuracy: {accuracy}")
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {accuracy}")
print(f"Attrition Accuracy: {accuracy}")
```
